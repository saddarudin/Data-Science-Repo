# It is used for categorical data

```python
from matplotlib import pyplot as plt

record = {"Saddar":77, "Azhar":87, "Faisal":45, "Fardeen":51, "Hassan":64}
names = list(record.keys())
roll_number = list(record.values())
```

```python
plt.bar(names,roll_number)
plt.title('Name-Roll Number Plot')
plt.xlabel('Names')
plt.ylabel('Roll Numbers')
plt.show()
```

**Create same plot but in horizontal**

```python
# since this is horizontal plot so position of x and y axis is changed
# this is why here x represents 'roll numbers' and y to 'names'
plt.barh(names,roll_number)
plt.title('Name-Roll Number Plot')
plt.xlabel('Roll Numbers')
plt.ylabel('Names')
plt.show()
```

# Scatter Plot

**In Scatter plot each point shows the difference between the values placed at the same index in the different data sets.**

```python
# both data sets must have same number of elements
x = [1,2,3,4,5,6]
y = [2,4,9,1,1,3]
plt.scatter(x,y)
plt.show()
```

**In the above graph change the dot into star**

```python
plt.scatter(x,y, marker='*')
plt.show()
```
