```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
flowers = load_iris()
dir(flowers)
```

```python
df = pd.DataFrame(flowers.data, columns=flowers.feature_names)
df.head()
```

```python
df['type']=flowers.target
df.head()
```

```python
flower=[]
for i in df.type:
  flower.append(flowers.target_names[i])
df['flower']=flower
df.head()
```

```python
df[df.type==2]
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

```python
kernel = ['linear', 'poly', 'rbf', 'sigmoid']
C = [1, 10, 20, 50, 100, 200]
net_score={}
for k in kernel:
  for c in C:
    scores = cross_val_score(SVC(kernel=k, C=c, gamma='auto'), flowers.data, flowers.target, cv=5)
    net_score[k+'_'+str(c)+': ']=scores

net_score
```

```python
net_score = pd.DataFrame(net_score)
net_score.head()
```

```python
net_score
```

# The above done work is all about manual but here is another method called GridSearch that does all the above work in single statement as shown below.

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = GridSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20,50,100],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
}, cv=5)

model.fit(flowers.data, flowers.target)
```

```python
pd.DataFrame(model.cv_results_)
```

```python
model.best_params_
```
