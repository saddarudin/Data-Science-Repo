**Create histograms for numeric variables to understand their distributions.**

```python
df.head()
plt.hist(df.year)
plt.xlabel('year')
plt.ylabel('selling_price')
plt.title('Selling Price Per Year')
plt.show()
```

**Use box plots to identify outliers and understand the spread of the data.**

```python
plt.boxplot(df.selling_price)
plt.ylabel('selling_price')
plt.title('Outliters in selling price')
plt.show()
```
