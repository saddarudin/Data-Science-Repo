# Line chart is used to show relationship between two numerical entities.

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,11)
y = 2*x
```

**Plot a simple line plot between x and y**

```python
plt.plot(x,y)
plt.show()
```

**Add title, x-label and y-label to the graph**

```python
plt.title('x-y relation')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.plot(x,y)
plt.show()
```

**Change color style and width of the line plotted on the graph**

```python
# different styles: "solid", "dashed", "dash dot", "dotted", "none", "-", "--", "-.", ":"
plt.plot(x,y, color='green', linestyle=":", linewidth=2)
plt.show()
```

**Adding two lines of y with one line of x. Also show grid lines in the background.**

```python
y2 = 3*x
plt.plot(y,x, color='red', linestyle="--", linewidth=2)
plt.plot(y2,x, color='yellow', linestyle="solid", linewidth=3)
# to show grid lines in the background
plt.grid(True)
plt.show()
```

**Make two subplots in a single row**

```python
plt.subplot(1,2,1)
# subplot(1,2,1) shows that two plots are on row 1 and in two columns last value
# shows index of the below
plt.plot(x,y,color='black',linestyle="--",linewidth=3)
plt.subplot(1,2,2)
plt.plot(x,y2,color='red',linestyle='-.',linewidth=2)
plt.show()
```
