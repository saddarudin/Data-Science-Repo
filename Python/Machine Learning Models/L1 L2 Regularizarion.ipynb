```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso # L1 Regularization
from sklearn.linear_model import Ridge # L2 Regularization
```

```python
df = pd.read_csv('Melbourne.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['Address', 'Date', 'Distance', 'Postcode', 'YearBuilt', 'Lattitude', 'Longtitude'], axis='columns')
df.shape
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.Price.fillna(df.Price.mean(), inplace=True)
df.isna().sum()
```

```python
df.Bedroom2
```

```python
df.Bedroom2.fillna(int(df.Bedroom2.mean()), inplace=True)
df.isna().sum()
```

```python
df.Bathroom.fillna(int(df.Bathroom.mean()), inplace=True)
df.isna().sum()
```

```python
df.Car.fillna(df.Car.median(), inplace=True)
df.isna().sum()
```

```python
df.Landsize.fillna(int(df.Landsize.mean()), inplace=True)
df.isna().sum()
```

```python
df.BuildingArea.fillna(int(df.BuildingArea.mean()), inplace=True)
df.isna().sum()
```

```python
df=df.dropna()
df.isna().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df.Suburb=le.fit_transform(df.Suburb)
df.head()
```

```python
df.Type = le.fit_transform(df.Type)
df.Method = le.fit_transform(df.Method)
df.SellerG = le.fit_transform(df.SellerG)
df.CouncilArea = le.fit_transform(df.CouncilArea)
df.Regionname = le.fit_transform(df.Regionname)
df.head()
```

```python
x = df.drop(['Price'], axis='columns')
y = df.Price
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)
x_test.shape
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

# Accuracy of model on testing data set

```python
model.score(x_test, y_test)
```

# Accuracy of model on training data set on which it is already trained.

```python
model.score(x_train, y_train)
```

# L1 Regularization

```python
model1 = Lasso()
model1.fit(x_train, y_train)
```

```python
print(f"Training Data Accuracy: {model1.score(x_train, y_train)}\nTesting Data Accuracy: {model1.score(x_test, y_test)}")
```

# L2 Regularization

```python
model2 = Ridge()
model2.fit(x_train, y_train)
```

```python
print(f"Training Data Accuracy: {model2.score(x_train, y_train)}\nTesting Data Accuracy: {model2.score(x_test, y_test)}")
```
