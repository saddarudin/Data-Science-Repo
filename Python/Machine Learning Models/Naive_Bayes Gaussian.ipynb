```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.drop(['PassengerId', 'Pclass', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns', inplace=True)
df.head()
```

```python
pd.isna(df).sum()
```

```python
df.Age.fillna(int(df.Age.mean()), inplace=True)
pd.isna(df).sum()
```

```python
df.head(6)
```

```python
sex=pd.get_dummies(df.Sex, dtype=int)
sex
```

```python
df = pd.concat([df,sex], axis='columns')
df
```

```python
df.drop(['Sex'], axis='columns', inplace=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(['Survived'], axis='columns'), df.Survived, test_size=0.20)
x_test.shape
```

```python
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test, y_test)
```
