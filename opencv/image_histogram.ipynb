```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### For zero channel image

```python
img = np.zeros((400, 400), np.uint8)
cv.rectangle(img,(100,50),(300,200),(128),-1)
cv.imshow('img',img)
hist = cv.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
cv.waitKey(0)
cv.destroyAllWindows()
```

### For colorful image/3 channel image

```python
img1 = cv.imread('resources/dog.jpg')
img1 = cv.resize(img1, (400, 400))
b,g,r = cv.split(img1)
cv.imshow('blue',b)
cv.imshow('green',g)
cv.imshow('red',r)
hist1 =plt.hist(b.ravel(),256,[0,256])
hist2 =plt.hist(g.ravel(),256,[0,256])
hist3 =plt.hist(r.ravel(),256,[0,256]) 
plt.title('colorful image')
plt.show()
cv.waitKey(0)
cv.destroyAllWindows()
```

### For gray_scale image


```python
img2 = cv.imread('resources/dog.jpg', 0)
img2 = cv.resize(img2, (400, 400))
cv.imshow('gray', img2)
gray_hist = cv.calcHist([img2],[0],None,[256],[0,256])
plt.plot(gray_hist)
plt.title('gray image')
plt.show()
cv.waitKey(0)
cv.destroyAllWindows()
```

### Histogram Equalization: It equalize the neighbor pixel values
### It works only on gray scale image

```python
equalized_img = cv.equalizeHist(img2)
result = np.hstack((img2,equalized_img))
cv.imshow('result',result)
equal_hist = cv.calcHist([equalized_img],[0],None,[256],[0,256])
plt.plot(equal_hist)
plt.title('equalized image')
plt.show()
cv.waitKey(0)
cv.destroyAllWindows()
```

### Contrast Limited Adaptive Histogram Equalization (CLAHE)
### Used to enhance image and also remove noise from it
### Requires gray scaled image

```python
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
# arguments are optional
cl1 = clahe.apply(img2)
cv.imshow('clahe',cl1)
clahe_hist = cv.calcHist([cl1],[0],None,[256],[0,256])
plt.plot(clahe_hist)
plt.title('clahe image')
plt.show()
cv.waitKey(0)
cv.destroyAllWindows()
```
