### morphological transformation basically has two types: erosion and dilation
### Erosion: it erodes away the boundaries of the foreground object
### Dilation: it increases the white region in the image or size of the foreground object
### Erosion works on black background and white objects

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('resources/dog.jpg',0)
thresh,binary = cv.threshold(img,127,255,cv.THRESH_BINARY)
kernel = np.ones((3,3),np.uint8)
```

## Erosion

```python
eroded = cv.erode(binary,kernel)
imgs = [img,eroded,binary]
titles = ['Original Image','Eroded Image','Binary Image']
plt.figure(figsize=(10,10))
for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(imgs[i],'gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

### Dilation

```python
dilated = cv.dilate(binary,kernel)
imgs = [img,dilated,binary,eroded]
titles = ['Original Image','Dilated Image','Binary Image','Eroded Image']
plt.figure(figsize=(6,6))
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(imgs[i],'gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

### Opening

```python
opening = cv.morphologyEx(binary,cv.MORPH_OPEN,kernel)
imgs = [img,binary,eroded,dilated,opening]
titles = ['Original Image','Binary Image','Eroded Image','Dilated Image','Opening Image']
plt.figure(figsize=(10,5))
for i in range(5):
    plt.subplot(2,3,i+1),plt.imshow(imgs[i],'gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

### Closing

```python
closing = cv.morphologyEx(binary,cv.MORPH_CLOSE,kernel)
imgs = [img,binary,eroded,dilated,opening,closing]
titles = ['Original Image','Binary Image','Eroded Image','Dilated Image','Opening Image','Closing Image']
plt.figure(figsize=(10,5))
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(imgs[i],'gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

### 1. Tophat 2. Gradient 3.Blackhat

```python
tophat = cv.morphologyEx(binary,cv.MORPH_TOPHAT,kernel)
gradient = cv.morphologyEx(binary,cv.MORPH_GRADIENT,kernel)
blackhat = cv.morphologyEx(binary,cv.MORPH_BLACKHAT,kernel)
imgs = [tophat,gradient,blackhat]
titles = ['Top Hat Image','Gradient Image','Black Hat Image']
plt.figure(figsize=(10,5))
for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(imgs[i],'gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
