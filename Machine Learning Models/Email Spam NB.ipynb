```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('spam.csv')
df.head()
```

```python
df['spam']=df.Category.apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
pipe = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.20)
x_test.shape
```

```python
pipe.fit(x_train, y_train)
```

```python
pipe.score(x_test, y_test)
```

```python
emails = [
    'Hi friend. Are you coming with me for the tour to Australia',
    'Congratulations! You have won 3,000,000 Rs. Contact on this number xxx'
]
```

```python
pipe.predict(emails)
```
