```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
dummies=pd.get_dummies(df.town, dtype=int)
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged.head(5)
```

```python
model = LinearRegression()
model.fit(merged[['area', 'monroe township', 'robinsville']], merged.price)
```

```python
model.predict([[2800, 0, 1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(merged[['area', 'monroe township', 'robinsville']], merged.price)
```

```python
with open('houseprices', 'wb') as f:
  pickle.dump(model, f)
```
