```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'])
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

It is necessary to save the trained model because you don't have to train the same model again and again on same data each time you run the code and the other benefit is you can share the binary file of trained model with your friend.

To save the any model there are two approaches:

1.   pickle
2.   joblib module of sklearn.external



```python
import pickle
```

```python
# I am opening file in wb mode it writes the file in binary format
with open('capita_income_model', 'wb') as f:
  pickle.dump(reg,f)
```

```python
with open('capita_income_model', 'rb') as f:
  my_model=pickle.load(f)
```

```python
my_model.predict([[2020]])
```

```python
import joblib
```

```python
joblib.dump(reg, 'trained_model')
```

```python
model=joblib.load('trained_model')
```

```python
model.predict([[2020]])
```
