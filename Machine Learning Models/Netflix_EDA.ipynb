```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('netflix_titles.csv')
df.head()
```

**What are the different types of content available on Netflix?**


```python
df.type.unique()
```

**Analyze the distribution of movies vs TV shows.**

```python
type = df.groupby('type')
type.describe()
```

**Which countries are the most common producers of Netflix content? Explore the distribution of content by country.**

```python
country_wise = df.groupby('country')
h=country_wise.release_year.describe()
h.sort_values('count', ascending=False).head()
```

**How has the number of titles added to Netflix changed over time? Plot the number of titles added each year or each month.**

```python
year_wise = df.groupby('release_year')
h=year_wise.title.describe()
h['count'].plot(kind='bar')
plt.xlabel('release_year')
plt.ylabel('count')
plt.title("Number of movies released over time")
plt.xticks(rotation=45)
plt.show()
# year_wise.title.plot(kind='bar', x='release_year')
# plt.show()
```
