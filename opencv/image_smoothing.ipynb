### Homogeneous Filter

#### This filter works like, each ouput pixel is the mean of its kernel neighbor
#### It is aka homogeneous filter in this all pixels contribute  with equal weight
#### In this filter kernel is [kernel/width*height]

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('resources/sdb.jpg',0)
cv.imshow('orignial image',img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
kernel = np.ones((2,2),np.float32)/4
h_filter = cv.filter2D(img,-1,kernel)
plt.imshow(h_filter,'gray')
plt.title('Homogeneous Filter')
plt.axis('off')
plt.show()
```

### Blur Filter
#### Takes the average of all the pixels under kernel area and replaces the central element with this average
#### Here kernel is save as it was for morphological transformation i.e (n,n)

```python
kernel = (5,5)
blur = cv.blur(img,kernel)
plt.imshow(blur,'gray')
plt.title('Blur Filter')
plt.axis('off')
plt.show()
```

### Gaussian Filter
#### The Gaussian filter works by convolving the image with a Gaussian kernel, a matrix that represents a Gaussian curve in 2D. Each pixel’s value is replaced by a weighted average of its neighbors, where closer neighbors are given more weight

```python
kernel = (3,3)
g_blur = cv.GaussianBlur(img,kernel,0)
plt.imshow(g_blur,'gray')
plt.title('Gaussian Blur Filter')
plt.axis('off')
plt.show()
```

### Median Filter
#### Computes the median of all the pixels under the kernel window and the central pixel is replaced with this median value.
#### This is highly effective in removing salt-and-pepper (white dots) noise.
#### Here kernel is a single value and it must be odd except 1

```python
m_blur = cv.medianBlur(img,3)
plt.imshow(m_blur,'gray')
plt.title('Median Blur Filter')
plt.axis('off')
plt.show()
```

### Bilaterial Filter
### It works like gaussian filter but more focus on edges
### It is slow as compated with other filters
### arguments(image,neighbor_pixel_diameter,sigma_color,sigma_space)

```python
bilateral = cv.bilateralFilter(img,3,75,57)
plt.imshow(bilateral,'gray')
plt.title('Bilateral Filter')
plt.axis('off')
plt.show()
```
