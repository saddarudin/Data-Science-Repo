```python
from sklearn.datasets import load_iris
```

```python
flower = load_iris()
dir(flower)
```

```python
flower.feature_names
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(flower.data, flower.target, test_size=0.20)
```

```python
len(x_test)
```

```python
model = LogisticRegression(solver='lbfgs',class_weight='balanced', max_iter=10000)
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_predicted = model.predict(x_test)
cm = confusion_matrix(y_test, y_predicted)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(cm, annot=True)
plt.title('Confusion Matrix for Iris Flowers Data Set')
plt.xlabel('predicted')
plt.ylabel('true')
plt.show()
```
