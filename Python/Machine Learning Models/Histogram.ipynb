**Histogram is used for continuous numerical columns.
On x-axis there is acutal value and on y-axis is the frequency of that value.**

```python
from matplotlib import pyplot as plt
```

```python
a = [2,2,3,4,2,7,8,2,7,6,8,8,5,9]
a
```

```python
plt.hist(a)
plt.show()
```

**Reduce number of bins in the histogram plotted above to the 4.**

```python
plt.hist(a,color='red',bins=4)
plt.show()
```

# Box plot
**It is used to show minimum, maximum and mean value in the data set.**

```python
l1 = [1,2,3,4,2,1,4,2,3,9]
plt.boxplot(l1)
plt.show()
```

**We can pass list of list in the argument and it will pring separate box for each list.**

```python
l2 = [2,1,2,4,5,7,8,6,8,9]
l3 = [1,9,4,8,6,7,3,2,4,8]
l=list([l1,l2,l3])
l
```

```python
plt.boxplot(l)
plt.show()
```

# Pie Chart
**To show the percentage of each category.**

```python
fruit = ['orange', 'banana', 'apple', 'mango', 'water melon']
quantity = [30, 60, 20, 40, 10]
```

```python
plt.pie(quantity, labels=fruit)
plt.show()
```

**Add percentage in each category in the above pie chart.**

```python
plt.pie(quantity, labels = fruit, autopct='%0.1f%%')
# 0.1f% means there should be one decimal place
plt.show()
```
