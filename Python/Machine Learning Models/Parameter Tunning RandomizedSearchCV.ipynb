```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
```

```python
flowers = load_iris()
dir(flowers)
```

```python
df = pd.DataFrame(flowers.data, columns=flowers.feature_names)
```

```python
df['target']=flowers.target
```

```python
df['target']=df.target.apply(lambda x: flowers.target_names[x])
df.head()
```

```python
model = RandomizedSearchCV(SVC(gamma='auto'),{
    'C':[1, 10, 100],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
}, cv=5)

model.fit(flowers.data, flowers.target)
```

```python
pd.DataFrame(model.cv_results_)
```

```python
model.best_score_
```
