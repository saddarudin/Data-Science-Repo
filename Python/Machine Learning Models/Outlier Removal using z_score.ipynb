```python
import pandas as pd
```

```python
df = pd.read_csv('heights.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title('Heights Plot')
plt.xlabel('height(inches)')
plt.ylabel('count')
plt.hist(df.height, bins = 20, rwidth=0.7)
plt.show()
```

```python
from scipy.stats import zscore
```

```python
z_score = zscore(df.height)
z_score
```

```python
df[(z_score<-3)| (z_score>3)]
```

```python
df=df[(z_score>-3) & (z_score<3)]
df.shape
```
